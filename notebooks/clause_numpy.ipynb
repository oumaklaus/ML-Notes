#MODULE 2 - Indexing & Slicing

```python
#Negative indexing
arr1 = np.random.randint(10,100, size = (3,3))
print(arr1)
```

```python
print(arr1[-1,-2]) #should print 44
print(arr1[-2,-3]) #should print 83
print(arr1[-1,-1]) #should print 40
print(arr1[-3,-2]) #should print 77
```

```python
#slicing
numbers = np.array([10,20,30,40,50,60,70,80,90])
print(numbers[1:4]) #should print from index 1 to 3. (1 is start : 4 is stop)
```

```python
#Ommiting values
print(numbers[3:]) #print numbers from the specifies index to the end
print(numbers[:3]) #print all numbers from beggining
```

```python
print(numbers[::2]) # start from index 0, then add 2 to the index and print the element of that index
print(numbers[::3]) # start from index 0, then add 3 to the index and print the element of that index
print(numbers[1::3]) # start from index 1, then add 3 to the index and print the element of that index
print(numbers[2:8:2]) # start from index 2, stop at index 8, add to 2, to the current index untill you reach the stop index
```

```python
#Reverse an Array

numbers[::-1]
```

```python
#Row Slicing
arr = np.array([
    [20,10,40],
    [39,20,45],
    [22,31,43]
])
print(arr[1]) #first row
print(arr[1]) #second row
print(arr[2]) #last row,, can also be written as print(arr[-1])
print(arr[1:]) #rows 1 and 2
```

```python
#Column Slicing

#arr[:, 0] #first column
arr[:,0:2] # print the first two columns
```
